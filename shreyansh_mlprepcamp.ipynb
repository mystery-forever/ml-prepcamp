```python
import numpy as np
import matplotlib.pyplot as plt

adj = 8
limits = [0, 100]

points = np.random.rand(10000, 2) * 100
grid = np.arange(limits[0], limits[1] + adj, adj)
grid_length = len(grid)
pair = set()

cell = [[[] for _ in range(grid_length - 1)] for _ in range(grid_length - 1)]

for i in range(len(points)):
    x = min(int(points[i, 0] / adj), grid_length - 2)
    y = min(int(points[i, 1] / adj), grid_length - 2)
    cell[x][y].append(i)

for i in range(grid_length - 2):
    for j in range(grid_length - 2):
        x0 = points[cell[i][j], 0]
        y0 = points[cell[i][j], 1]
        plt.scatter(x0, y0, s=1)
```
